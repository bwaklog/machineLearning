```python
from tensorflow import keras
import numpy as np
```

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# list of x values
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
# list of y values
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
# Training takes place with .fit command
# We are asking the model to figure out how to fit x vals to y vals
# epochs --> no. of iterations through training loop

# Training loop :
#    It makes a guess, measures how good or bad the guess is with the
# loss function, then uses the optimizer and the data to make another
# guess and repeat this.
```

```python
model.fit(xs, ys, epochs=500)
```

```python
# Now we ask the model to predict the value of x for an unknown value
print(model.predict([10.0]))
# The number is very close to 19, but not 19
# This is because we have trained it will very less points (6 points)
```
